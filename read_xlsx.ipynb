```python
import pandas as pd
import xlrd
from openpyxl import load_workbook
```

```python
# 读文件
workbook = load_workbook('Timetable.xlsx')

allsheets = workbook.sheetnames
subway_list = []

for sheet_i in range(0,len(allsheets)):
    sheetname = allsheets[sheet_i]
    line_i = 16 if (sheet_i==13) else (sheet_i+1)
    sheet = workbook[sheetname]
    rows = sheet.max_row
    node_list = []
    if (sheet_i == 9) or (sheet_i == 10): # 10号线和11号线，因为有两个方向，header占了两行，要从第三行开始，跳过这一次循环
        row_start = 3
    else:
        row_start = 2
    for row in range(row_start,rows+1):#从第2行开始取数据
        if (sheet_i == 9) or (sheet_i == 10): # line 10 , line 11
            node_list.append(['line '+str(line_i),sheet[row][0].value,sheet[row][1].value,sheet[row][2].value])
        else:
            node_list.append(['line '+str(line_i),sheet[str(row)][0].value,sheet[str(row)][1].value])
    #print(node_list)
    subway_list.append(node_list)

#print(subway_list[0][0][0]) # 莘庄

# 处理数据中的\xa0空格
for line in subway_list:
    for station in line:
        #print(type(station[1]))
        station[1] = ''.join(station[1].split())
        print(station[1],0) # 加0是为了方便看空格在后面的情况
```

```python
#处理datetime
import datetime
#start_time = datetime.time(21,30)
#minutes2 = start_time.hour * 60 + start_time.minute

for i in range(0, len(subway_list)):
    start_time = subway_list[i][0][2]
    minutes2 = start_time.hour * 60 + start_time.minute
    for station in subway_list[i]:  # 去掉第一行
        if i == 9: # 10号线分叉，对应的时间有两列
            if station[2] == '---': # 第三列， 开往虹桥火车站方向. 且起始时间晚了5分钟,要减去
                tm = station[3]
                minutes1 = tm.hour * 60 + tm.minute - 5
            else :
                tm = station[2]
                minutes1 = tm.hour * 60 + tm.minute
            station.pop()
        elif i == 10: # 11号线分叉，对应的时间有两列
            if station[2] == '---': # 第三列， 开往花桥方向. 且起始时间比前一列早了30分钟分钟,要加上
                tm = station[3]
                minutes1 = tm.hour * 60 + tm.minute + 30
            else :
                tm = station[2]
                minutes1 = tm.hour * 60 + tm.minute
            station.pop()
        else :
            tm = station[2]
            minutes1 = tm.hour * 60 + tm.minute
        # 所有条件都要执行的
        minutes = minutes1 - minutes2
        station[2] = minutes
        print(station)
```

```python
# 写文件
file_handler =open('subway.txt',mode='w',encoding='gbk')
#写入数据
for line in subway_list:
    for station in line:
        s = ' '.join(str(i) for i in station) 
        file_handler.write(s)
        file_handler.write('\n')
file_handler.close()
```
